```python
import matplotlib.pyplot as plt
import numpy as np

# 資料
x = np.array([2, 3, 4, 5])  # X軸
ours_y = np.array([0.013, 0.0051, 0.0029, 0.0022])
multiHist_y = np.array([0.020, 0.0067, 0.0037, 0.0026])
copulaBase_y = np.array([0.018, 0.0072, 0.0050, 0.0045])

# 將它們組成一個矩陣方便操作 (3 methods × 4 variables)
data = np.vstack([ours_y, multiHist_y, copulaBase_y])

# 對每個 x (每一欄) normalize 到 [0,1]
normalized = data / (data.sum(axis=0) + 1e-8)

# 分別取出 normalize 後的結果
ours_norm, multiHist_norm, copulaBase_norm = normalized

# 畫線圖
plt.figure(figsize=(8, 5))
plt.plot(x, ours_norm, marker='o', label='Ours')
plt.plot(x, multiHist_norm, marker='s', label='multi-histogram')
plt.plot(x, copulaBase_norm, marker='^', label='Copula Base')

# 標籤
plt.xlabel('Number of Variables')
plt.ylabel('Normalized RMSE')
plt.title('Normalized Comparison of Methods')

plt.xticks(x)
plt.legend()
plt.grid(True, linestyle='--', alpha=0.6)
plt.tight_layout()
plt.show()
```
