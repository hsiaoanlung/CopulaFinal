```python
import matplotlib.pyplot as plt
import numpy as np

# 
x = np.array([2, 3, 4, 5])  # X
ours_y = [477864, 565704, 666612, 757080]
### block 3 ###
multiHist_y = [904842, 1846460, 3752975, 5966862]
### block 2 ###
# multiHist_y = [2692449, 5063888, 9455265, 14398170]
copulaBase_y = [2162688, 2647654, 3198156, 3742105]

# 
plt.figure(figsize=(8, 5))
plt.plot(x, ours_y, marker='o', label='Ours')
plt.plot(x, multiHist_y, marker='s', label='multi-histogram')
plt.plot(x, copulaBase_y, marker='^', label='Copula Base')

# 
plt.xlabel('Number of Variables')
plt.ylabel('Storage')
plt.title('Comparison of Storage')

# x
plt.xticks(x)
#plt.yscale('log')
# 
plt.legend()
plt.grid(True, linestyle='--', alpha=0.6)

plt.tight_layout()
plt.show()
```
